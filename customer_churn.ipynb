
# **Common characteristics of high-risk churn customers:**


Mostly on **Basic subscription with Monthly contracts** (highest churn rate)

Have **higher tenure**, indicating **pricing fatigue or competitor switching**

Make **more support calls**, suggesting **unresolved issues and dissatisfaction**

Experience more **payment delays, increasing churn** likelihood

Show lower engagement / usage frequency

Have longer gaps since last interaction, **indicating reduced interest**

Contribute to significant **revenue loss** despite churning

**Female customers** show a relatively **higher churn share**

Average spend differs, but churned users tend to **reduce long-term value**

Likely motivated by **better competitor offers or unmet expectations**
